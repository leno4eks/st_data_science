Отбрасывание записей и признаков . Это простое удаление записей или признаков в которых содержаься пропуски\
"Здесь важно правильно выбрать ось удаления: если мы избавимся от большого числа строк, то рискуем потерять важные данные, а если мы удалим столбцы, то можем потерять важные признаки."


Прежде всего порассуждаем логически: в столбце hospital_beds_raion более 47 % пропусков. Если мы будем удалять все строки, в которых этот признак пропущен, мы потеряем почти половину наших данных! 

Правильнее будет просто удалить столбец, ведь число мест в районных больницах — это не самый информативный признак для определения цены квартиры.

А вот если мы удалим весь столбец metro_km_walk, где менее 1 % пропусков, то потеряем полезный признак при формировании прогноза цены, ведь расстояние до ближайшего метро — это важный фактор при выборе квартиры. В данном случае лучше будет удалить сами записи.

рекомендуют при использовании метода удаления придерживаться следующих правил: удаляйте столбец, если число пропусков в нем более 30-40 %. В остальных случаях лучше удалять записи.

**Для удаления столбцов используем метод dropna()**\
*axis* -ось по которой производится удаление( по умолчанию 0- строки)\
*how*-  как производится удаление строк (any- еслт хотябы бы в одном из столбцов пропуск( стоит по умолчанию), если во всех  столбцах  есть пропуски)\
*thresh* -порог удаления . определяет мин число непустых значений в строке столбце ,  при котором она/он сохраняется. например если естановить в его значении 2, то мы удалим строки где число пропусков больше n-2,  где n число признаков 

```python
#создаем копию исходной таблицы
drop_data = sber_data.copy()
#создаем минимальный порог: вычисляем 70% от числа пропусков 
thresh= drop_data.shape[0]*0.7
# удаляем столбцы, в которых пропущено более 30% пропусков 
drop_data = drop_data.dropna(thresh=thresh, axis=1)
#удаляем столбца где есть хоть один пропуск 
drop_data=drop_data.dropna(how = 'any', axis =0)
# отображаем результирующую долю пропусков 
drop_data.isnull().mean()
```

заполнение значений осуществляется при помощи метода  fillna()
Главный параметр метода value(значение на которое происходит заполнение)
Если метод вызывается от имени всего data_frame  то в качестве value можно использовать словарь , где ключи -названия слотбцов, значения- константы 

```python
#создаем копию таблицы
fill_data=sber_data.copy()
#создаем словарь имя столбца:число(признак) на которые надо заменить пропуски

values ={
    'life_sq':fill_data['full_sq'],
    'metro_min_walk': fill_data['metro_min_walk'].median(),
    'metro_km_walk': fill_data['metro_km_walk'].median(),
    'railroad_station_walk_km':fill_data['railroad_station_walk_km'].median(),
    'railroad_station_walk_min': fill_data['railroad_station_walk_min'].median(),
    'hospital_beds_raion': fill_data['hospital_beds_raion'].mode()[0],
    'preschool_quota':fill_data['preschool_quota'].mode()[0],
    'school_quota':fill_data['school_quota'].mode()[0],
    'floor': fill_data['floor'].mode([0])
}

#заполняем пропуски в соответсвие с заявленнным словарем
fill_data=fill_data.fillna(values)
#выводим результатирующую долю пропусков 
fill_data.isnull().mean()
```

**Заполнение недостающих  значений констант с добавлением индикатора**

То есть создаем новые столбцы , где будет указывать было ли изменение в таблице или нет, создано для того чтобы уменьшить искажение, но если пропусков много лучше не использовать , так как задублируется множество столбцов 

Посмотрим на реализацию. Как обычно, создадим копию indicator_data исходной таблицы. В цикле пройдёмся по столбцам с пропусками и будем добавлять в таблицу новый признак (с припиской "was_null"), который получается из исходного с помощью применения метода isnull(). После чего произведём обычное заполнение пропусков, которое мы совершали ранее, и выведем на экран число отсутствующих значений в столбце, чтобы убедиться в результате:

```python
#создаем копию исходной таблицы
indicator_data = sber_data.copy()
#в цикле пробегаемся по названиям столбцов с пропусками
for col in cols_with_null.index:
    #создаем новый признак-индикатор как col_was_null
    indicator_data[col + '_was_null'] = indicator_data[col].isnull()
#создаем словарь имя столбца: число(признак) на который надо заменить пропуски   
values = {
    'life_sq': indicator_data['full_sq'],
    'metro_min_walk': indicator_data['metro_min_walk'].median(),
    'metro_km_walk': indicator_data['metro_km_walk'].median(),
    'railroad_station_walk_km': indicator_data['railroad_station_walk_km'].median(),
    'railroad_station_walk_min': indicator_data['railroad_station_walk_min'].median(),
    'hospital_beds_raion': indicator_data['hospital_beds_raion'].mode()[0],
    'preschool_quota': indicator_data['preschool_quota'].mode()[0],
    'school_quota': indicator_data['school_quota'].mode()[0],
    'floor': indicator_data['floor'].mode()[0]
}
#заполняем пропуски в соответствии с заявленным словарем
indicator_data = indicator_data.fillna(values)
#выводим результирующую долю пропусков
indicator_data.isnull().mean()
```

**Комбинирование методов**\
К примеру:
1. Удалить столбцы в которых пропусков больше 30%\
2. Удалить записи, в которых  более 2 пропусков одновременно\
3. Заполнить остаток константами

```python
#создаем копию исходной таблицы
combine_data= sber_data.copy()

#отбрасываем столбцы с числом пропусков больше 30% (100-70)
n=combine_data.shape[0] #число строк в таблице
thresh=n*0.7
combine_data= combine_data.dropna(thresh=thresh,axis=1)

#отбрасываем строки с числом пропусков более 2 в строке 
m=combine_data.shape[1] #число признаков после удаления столбцов
combine_data = combine_data.dropna(thresh=m-2,axis=0)

#создаем словарь "имя столбца ": число ( признак), на который надо зааменить пропуски
values = {
    'life_sq': combine_data['full_sq'],
    'metro_min_walk': combine_data['metro_min_walk'].median(),
    'metro_km_walk': combine_data['metro_km_walk'].median(),
    'railroad_station_walk_km': combine_data['railroad_station_walk_km'].median(),
    'railroad_station_walk_min': combine_data['railroad_station_walk_min'].median(),
    'preschool_quota': combine_data['preschool_quota'].mode()[0],
    'school_quota': combine_data['school_quota'].mode()[0],
    'floor': combine_data['floor'].mode()[0]
}

#заполняем оставшиеся пропуски константами в соответсвии со словарем

combine_data=combine_data.fillna(values)
display(combine_data.isnull().mean())
```

```python
#выведем результируюющее число строк и столбцов
print(combine_data.shape)
```
