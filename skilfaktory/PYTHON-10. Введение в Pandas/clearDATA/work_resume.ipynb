```python
def clean_data(data):
    """
    Функция для очистки данных от специальных символов.

    Args:
        data (str): Данные, которые нужно очистить.

    Returns:
        str: Очищенные данные.
    """

    # Удаляем специальные символы из строки
    data = data.replace('_', '')
    data = data.replace('~','')
    data = data.replace('\\','')
    data = data.replace('/','')
    data = data.replace('-','')

    return data

# Применяем функцию к столбцам producer_id и producer_color
for i in range(len(Data_Parsing)):
    Data_Parsing['producer_id'][i] = clean_data(Data_Parsing['producer_id'][i])
    Data_Parsing['producer_color'][i] = clean_data(Data_Parsing['producer_color'][i])

# Выводим первые 15 строк Data_Parsing
Data_Parsing.head(15)
```
